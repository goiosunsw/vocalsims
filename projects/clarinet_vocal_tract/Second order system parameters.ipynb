# Second order system parameters

A second order system can be put into the form

$$ \frac{d^2 y}{dt^2} + \frac{\omega_0}{Q_0} \frac{d y}{dt} + \omega_0^2 y = b_1 \frac{d x}{dt} + b_0 x$$

Where $x(t)$ is the forcing input and $y(t)$ is the system's response

## Laplace transform

The particular solution of these systems to a unit pulse at $t=0$ can be put in the form 

$$ y_p(t) = y_0 e^{s_0 t} u(t) $$ 

Where $s_0$ is a complex number and $u(t)$ is the unit step function $u(t) = 0$ for $t<0$ and $u(t) = 1$ for $t>1$ . This is the solution for a unit pulse at $t=0$ 

For a general input $x(t)$ the response will be a sum of different $s$ values

$$ y(t) = \int Y(s) e^{s t} ds $$

### Response in Laplace form

Because any function $f(t)$ can be decomposed in a sum of decaying exponentials, both $x(t)$ and $y(t)$ can be written as in the last equation. $Y(s)$ is then the Laplace transform of $y(t)$ 

Also, since the differential equation is a linear system, knowing the response for all values of $s$ means that we can calculate the response to any input $x(t)$ by integrating over $s$ 

The Laplace response is obtained by replacing $y(t) = Y(s) e^{s t}$ and $x(t) = X(s) e^{s t} $

$$ s^2 Y + s \frac{\omega_0}{Q_0} Y + \omega_0^2 Y = b_1 s X + b_0 X$$

so that 

$$ Y = \frac{b_1 s  + b_0 }{s^2  + s \frac{\omega_0}{Q_0}  + \omega_0^2 } X $$

## Solution for unit pulse at $t=0$

This is the free-oscillating solution

Replace $y(t) \rightarrow y_p(t)$ in the differential equation

$$ s_0^2 y_0 e^{s_0 t}  + \frac{\omega_0}{Q_0} s_0 y_0 e^{s_0 t}  + \omega_0^2  y_0 e^{s_0 t}  = 0$$

for $t>0$, or

$$ s_0^2  + \frac{\omega_0}{Q_0} s_0  + \omega_0^2    = 0$$

so that

$$ s_0 = -\frac{\omega_0}{2 Q_0} \pm \sqrt{\left(\frac{\omega_0}{2 Q_0}\right)^2-\omega_0^2} $$

or 

$$ s_0 = \omega_0 \left( \pm \sqrt{\frac{1}{4 Q_0^2}-1} - \frac{1}{2 Q_0}\right) $$

### When $Q_0$ is large

$$ s_0 = \pm \jmath  \omega_f - \frac{\omega_0}{2 Q_0}$$

with $\omega_f = \omega_0 \sqrt{1-\frac{1}{4 Q_0^2}}$, close to $\omega_0$ for large values of $Q_0$

and the solution is 

$$ y_p(t) = e^{- \frac{\omega_0}{2 Q_0} t} \left(y_0^+  e^{+\jmath\omega_f t}  + y_0^-  e^{-\jmath\omega_f t} \right)u(t)$$ 


### Poles

The two solutions of $s_0^+ = s_0$ and $s_0^- =s_0^*$ are the poles of the response function of the system because they make its denominator $=0$

## Amplitude of response to a unit pulse

The Laplace transform of a unit pulse $x(t) = \delta(t)$ is $X(s) = 1$ 

The Laplace transform of the response is then given by $Y$ by replacing $X=1$

$$ Y_f = \frac{b_1 s  + b_0 }{s^2  + s \frac{\omega_0}{Q_0}  + \omega_0^2 } $$

The response is the inverse Laplace transform of $Y_f$.



### Known Laplace transforms that can be useful in our case

| $f(t)$ | $F(s)$ |
| :-----: | :-------: |
| $\delta(t)$ | $1$ |
| $e^{s_0 t}$ | $\frac{1}{s+s_0}$ |
| $a f(t)$ | $a F(s)$ |
| $f(t) + g(t) $ | $F(s)+G(s)$ |


The fraction of polynomial functions in $Y_f$ can be represented as a linear superposition of Laplace transforms of $e^{s_0 t}$.

$$ 
\begin{align}
Y_f & = \frac{b_1 s  + b_0 }{s^2  + s \frac{\omega_0}{Q_0}  + \omega_0^2 }  \\
    & = \frac{c_1}{s+s_0} + \frac{c_2}{s+s_0^*} \\
    & = \frac{c_1(s+s_0^*)+ c_2(s+s_0)}{s^2+2 \Re{(s_0)} s + |s_0|^2} 
\end{align}
$$

The denominator just tells us what we already know, the solutions for $s_0$ found above. 

The numerator has to be determined independently as a function of the values of $b_1$ and $b_0$. We have

$$ 
b_1 s + b_0  = (c_1+c_2) s + (c_1 s_0^* + c_2 s_0)
$$

so that 
$$
\begin{align}
b_1 &=& c_1 &+& c_2 \\
b_0 &=& c_1 s_0^* &+& c_2 s_0
\end{align}
$$

$$
\begin{align}
- b_1 s_0 &= - c_1 s_0  &-& c_2 s_0  \\
b_0 &= c_1 s_0^* &+& c_2 s_0
\end{align}
$$

$$
b_0 - b_1 s_0  = c_1 (s_0^* - s_0) 
$$

$$
\begin{align}
c_1  &=  \frac{b_0 - b_1 s_0 }{s_0^* - s_0} \\
     &=  \frac{b_0 - b_1 s_0 }{-2\jmath\Im(s_0)} \\ 
     &=  \jmath \frac{b_0 - b_1 s_0 }{2\Im(s_0)} \\
     &=  \jmath \frac{b_0 - b_1 \Re(s_0) - \jmath b_1 \Im(s_0) }{2\Im(s_0)} \\
     &=  \frac{  b_1 \Im(s_0) + \jmath(b_0 - b_1 \Re(s_0))  }{2\Im(s_0)} \\
     &=    b_1  + \jmath\frac{(b_0 - b_1 \Re(s_0))  }{2 \Im(s_0)}
\end{align}
$$

$$
\begin{align}
b_0 - b_1 s_0^*  &= c_2 (s_0 - s_0^*) \\
c_2  &=  -\jmath \frac{b_0 - b_1 s_0^* } {2 \Im(s_0)} \\
     &=  \frac{  b_1 \Im(s_0) - \jmath(b_0 - b_1 \Re(s_0))  }{ 2 \Im(s_0)} \\
     &= c_1^*
\end{align}
$$

Usually either $b_0$ or $b_1$ is $0$

### $b_1 = 0$

$$c_1 = -c_2$$

$$ c_1 (s_0 - s_0^*) = b_0 $$

$$ c_1  = \frac{b_0}{\jmath\Im{s_0}} $$
$$ c_2  = -\frac{b_0}{\jmath\Im{s_0}} $$

### $b_0 = 0$

$$ c_2 = b_1-c_1$$

$$ c_1 s_0 + c_2 s_0^* = 0  $$

$$ c_1 s_0 + (b_1-c_1) s_0^* = 0 $$

$$ c_1 (s_0 - s_0^* ) = -b_1 s_0^* $$ 

$$ c_1  = -b_1 \frac{s_0^*}{\jmath\Im{s_0}} $$ 

----
$$ 
\begin{align}
c_2  &= b_1\left(1+ \frac{s_0^*}{\jmath\Im{s_0}}\right) \\
    &=   b_1\left(\frac{s_0 - s_0^* + s_0^*}{s_0 - s_0^*}\right) \\
    &=   b_1\left(\frac{s_0 }{s_0 - s_0^*}\right) \\
    &=   b_1\left(\frac{s_0 }{\jmath\Im{s_0}}\right) \\
    &= c_1^*
\end{align}$$ 

----

$$ 
\begin{align}
Y_f & = \frac{b_1 s  + b_0 }{s^2  + s \frac{\omega_0}{Q_0}  + \omega_0^2 }  \\
    & =  \left( b_1  + \jmath\frac{(b_0 - b_1 \Re(s_0))  }{2 \Im(s_0)}\right)\frac{1}{s+s_0} +
         \left( b_1  - \jmath\frac{(b_0 - b_1 \Re(s_0))  }{2 \Im(s_0)}\right)\frac{1}{s+s_0^*} 
\end{align}
$$


so the solution in time domain is:

$$
\begin{align}
y(t) & = \left( b_1  + \jmath\frac{(b_0 - b_1 \Re(s_0))  }{2 \Im(s_0)}\right) e^{s_0 t} &+ 
       \left( b_1  + \jmath\frac{(b_0 - b_1 \Re(s_0))  }{ 2 \Im(s_0)}\right) e^{-s_0 t}
     & = e^{\Re(s_0)}  
\end{align}
$$
